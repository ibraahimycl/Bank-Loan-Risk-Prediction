```python
bureau_agg = fd.groupby("SK_ID_CURR").agg(
    NUM_CREDITS=("SK_ID_BUREAU", "count"),
    NUM_ACTIVE=("CREDIT_ACTIVE", lambda x: (x == "Active").sum()),
    NUM_CLOSED=("CREDIT_ACTIVE", lambda x: (x == "Closed").sum()),
    NUM_SOLD=("CREDIT_ACTIVE", lambda x: (x == "Sold").sum()),
    NUM_BADDEBT=("CREDIT_ACTIVE", lambda x: (x == "Bad debt").sum()),
    TOTAL_DEBT=("AMT_CREDIT_SUM_DEBT", "sum"),
    TOTAL_CREDIT=("AMT_CREDIT_SUM", "sum"),
    TOTAL_OVERDUE=("AMT_CREDIT_SUM_OVERDUE", "sum"),
    AVG_DAYS_CREDIT=("DAYS_CREDIT", "mean"),
    MAX_OVERDUE=("CREDIT_DAY_OVERDUE", "max")
).reset_index()

bureau_agg["DEBT_RATIO"] = bureau_agg["TOTAL_DEBT"] / bureau_agg["TOTAL_CREDIT"]
bureau_agg["RISKY_CREDIT_RATIO"] = (bureau_agg["NUM_BADDEBT"] + bureau_agg["NUM_SOLD"]) / bureau_agg["NUM_CREDITS"]

bureau_agg.head()
```

```python
fd["IS_RISKY_TYPE"] = fd["CREDIT_TYPE"].isin(["Consumer credit", "Credit card"])
fd["IS_SECURED_TYPE"] = fd["CREDIT_TYPE"].isin(["Mortgage", "Car loan"])

bureau_type_agg = fd.groupby("SK_ID_CURR").agg(
    NUM_RISKY_CREDITS=("IS_RISKY_TYPE", "sum"),
    NUM_SECURED_CREDITS=("IS_SECURED_TYPE", "sum")
).reset_index()

bureau_agg = bureau_agg.merge(bureau_type_agg, on="SK_ID_CURR", how="left")
```

```python
fd["RECENT_CREDIT"] = (fd["DAYS_CREDIT"] >= -365).astype(int)
fd["OLD_CREDIT"] = (fd["DAYS_CREDIT"] <= -1825).astype(int)

time_agg = fd.groupby("SK_ID_CURR").agg(
    RECENT_CREDIT_COUNT=("RECENT_CREDIT", "sum"),
    OLD_CREDIT_COUNT=("OLD_CREDIT", "sum")
).reset_index()

bureau_agg = bureau_agg.merge(time_agg, on="SK_ID_CURR", how="left")
```

```python
fd["HAS_OVERDUE"] = (fd["CREDIT_DAY_OVERDUE"] > 0).astype(int)
fd["LONG_OVERDUE"] = (fd["CREDIT_DAY_OVERDUE"] > 90).astype(int)

overdue_agg = fd.groupby("SK_ID_CURR").agg(
    HAS_OVERDUE_FLAG=("HAS_OVERDUE", "max"),
    LONG_OVERDUE_COUNT=("LONG_OVERDUE", "sum")
).reset_index()

bureau_agg = bureau_agg.merge(overdue_agg, on="SK_ID_CURR", how="left")
```

```python
print("Final Bureau Aggregated Features (Customer-Level):")
print(bureau_agg.sample())
print(f"\nShape: {bureau_agg.shape}")
```
